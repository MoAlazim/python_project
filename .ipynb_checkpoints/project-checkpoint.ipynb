```python
'''
This program allows one to search through the images looking for the occurrences of keywords and faces. E.g. if you search 
for "pizza" it will return a contact sheet of all of the faces which were located on the images which mentions "pizza". 
'''
from zipfile import ZipFile
import PIL
import math
from PIL import Image, ImageDraw, ImageFont
import pytesseract
import cv2 as cv
import numpy as np

face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')


def getImageFiles():
    '''
    Extract all image files from the zip file and return each image with it's name
    '''
    images = []
    files = ZipFile('readonly/small_img.zip')
    for file in files.filelist:
        img = Image.open(files.extract(file.filename))
        images.append((img, file.filename))
    return images


def resizer(img):
    ''' Resize and return images in same size, you have to change [basewidth] to the size you want'''
    basewidth = 160 
    wpercent = (basewidth / float(img.size[0]))
    hsize = int((float(img.size[1]) * float(wpercent)))
    img = img.resize((basewidth, hsize), PIL.Image.ANTIALIAS)
    return img


def contactSheet(data):
    '''
    - Creat a contact sheet for every image faces, then convert the image from CV2 format to PIL format
    - add faces to contact sheet and display it
    '''
    font = ImageFont.truetype("readonly/fanwood-webfont.ttf", 30)
    newData = []
    
    for fileName, faces in data:
        pil_faces = []
        for face in faces:
            img = Image.fromarray(face)
            pil_faces.append(img)
        newData.append((fileName, pil_faces))

    first_image = newData[0][1][0]
    first_image = resizer(first_image)
    
    for fileName, faces in newData:
        x=0
        y=50
        if len(faces) < 5:
            coef = 1
            contact_sheet=PIL.Image.new(first_image.mode, (int(first_image.width*5), int(first_image.height*coef)))
        else:
            coef = int(math.ceil(len(faces)/5))
            contact_sheet=PIL.Image.new(first_image.mode, (int(first_image.width*5), int(first_image.height*coef)))
        
        obj = ImageDraw.Draw(contact_sheet)
        obj.text((10,10), '({}) Face/s | Result found in File: ({})'.format(len(faces), fileName), fill=(255,255,255), font=font)
        for face in faces:
            contact_sheet.paste(resizer(face), (x, y) )
            if x+first_image.width == contact_sheet.width:
                x=0
                y=y+first_image.height
            else:
                x=x+first_image.width

        display(contact_sheet)


def extractText(images):
    ''' Return extracted text from the image '''
    extracted = []
    for img, fileName in images:
        im = img.convert('L')
        text = pytesseract.image_to_string(im)
        extracted.append((fileName, img, text))
    return extracted


def search(text, keyword):
    ''' Return True and file name if the entered keyword was found, else return False '''
    if keyword in text or keyword.lower() in text or keyword.upper() in text or keywork.title() in text:
        return True


def faceDetection(img):
    ''' Return image file in CV2 format with faces detected'''
    img.save('image.png')
    im = cv.imread('image.png')
    gray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.3,
            minNeighbors=3,
            minSize=(30, 30)
            )
    return (im, faces)


def cropFaces(data):
    ''' Return detected faces on the image after cropped them '''
    newResult = []
    for fileName, img, faces in data:
        croped_faces = []
        for (x, y, w, h) in faces:
            cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            face = img[y:y + h, x:x + w]
            croped_faces.append(face)
        newResult.append((fileName, croped_faces))
    return newResult


def showRects(data):
    ''' Do not waste your time on reviewing this function, it is just for fun, it drawing a rectangle  around
        detected faces on the image
    '''
    for fileName, img, faces in data:
        drawing=ImageDraw.Draw(img)
        for x,y,w,h in faces:
            drawing.rectangle((x,y,x+w,y+h), outline="red")
        display(img)


def mainScript():
    ''' Ask the user to enter a keyword to search for in the images, then return the results
        as a contact sheet of faces that found on the images
    '''
    found = False
    result = []
    keyword = input('Please enter a keyword: ')
    try:
        if isinstance(keyword, str):
            textInfo = extractText(getImageFiles())
            for fileName, img, text in textInfo:
                found = search(text, keyword)
                if found:
                    faces = faceDetection(img)
                    image, facesInImage = faces
                    if len(facesInImage) != 0:
                        result.append((fileName, image, facesInImage))
                    else:
                        print('though we found result for your keyword, in the file: {}, we did not find any faces associated.'.format(fileName))
                else:
                    print('No result found in the file: {}.'.format(fileName))
        else:
            print('Please enter just a text')
        contactSheet(cropFaces(result))
    except Exception as e:
        print('Something went wrong, Please try again.')
        print(e)


if __name__ == '__main__':
    mainScript()
```
